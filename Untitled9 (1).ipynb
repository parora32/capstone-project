```python
import pandas as pd            
import numpy as np             
import matplotlib.pyplot as plt 
import seaborn as sns          
```

```python
df = pd.read_csv("Dentistry_Dataset_new.csv")
```

```python
df.head()
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)


```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['ICDI'] = label_encoder.fit_transform(df['ICDI'])
```

```python
# Assuming 'Age' is the name of the dependent variable column
X = df.drop('ICDI', axis=1)
Y = df['ICDI']
```

```python
from sklearn.preprocessing import Normalizer

# Initialize the normalizer
normalizer = Normalizer()

# Fit and transform the data
X_normalized = normalizer.fit_transform(X)
```

```python
# Correlation matrix
corr_matrix = df.corr()

# Heatmap using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Drop columns that are not important or are highly correlated
X_dropped = X.drop(['LCII', 'LCIC'], axis=1)
```

```python
from sklearn.model_selection import train_test_split

# Split the data
X_train, X_test, Y_train, Y_test = train_test_split(X_dropped, Y, test_size=0.2, random_state=42)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Initialize and train the model
log_reg = LogisticRegression()
log_reg.fit(X_train, Y_train)

# Predict and evaluate
Y_pred_log_reg = log_reg.predict(X_test)
accuracy_log_reg = accuracy_score(Y_test, Y_pred_log_reg)
print(f'Logistic Regression Accuracy: {accuracy_log_reg}')
```

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize and train the model
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train, Y_train)
Y_pred_dt = dt_classifier.predict(X_test)
accuracy_dt = accuracy_score(Y_test, Y_pred_dt)
print(f'Decision Tree Classifier Accuracy: {accuracy_dt}')
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize and train the model
rf_classifier = RandomForestClassifier()
rf_classifier.fit(X_train, Y_train)
Y_pred_rf = rf_classifier.predict(X_test)
accuracy_rf = accuracy_score(Y_test, Y_pred_rf)
print(f'Random Forest Classifier Accuracy: {accuracy_rf}')
```
